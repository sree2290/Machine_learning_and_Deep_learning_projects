```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data  = pd.read_csv("MNIST.csv")
```

```python
data
```

```python
data["Index"].value_counts()
```

```python
data.describe()
```

```python
sns.countplot(data.Index);
```

#### No imbalanced Data so we can use random sampling

###  Spliting X and y 

```python
X = data.loc[:, data.columns != 'Index']
y = data["Index"]
```

```python
X
```

```python
y
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=69)
```

```python
print(X_train.shape,y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
sns.countplot(y_train)
```

```python
sns.countplot(y_test)
```

###  Logistic

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train, y_train)
```

```python
r_y_pred = reg.predict(X_test)
```

```python
log_df = pd.DataFrame({'Actual': y_test, 'Predicted': r_y_pred})
log_df.head()
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(r_y_pred,y_test)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train, y_train)
```

```python
y_test_pred_knn = knn_classifier.predict(X_test)
```

```python
knn_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred_knn})
knn_df.head()
```

```python
metrics.accuracy_score(y_test_pred_knn,y_test)
```

### SVM

```python
from sklearn.svm import SVC
sv_classifier = SVC()
sv_classifier.fit(X_train, y_train)
```

```python
y_test_pred_svc = sv_classifier.predict(X_test)
```

```python
metrics.accuracy_score(y_test_pred_svc,y_test)
```

```python
svm_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred_svc})
svm_df.head()
```
