Trafic_sign_prediction

```python
from keras.models import load_model
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the model
model = load_model('/content/Trafic_sign_Pred_model.h5')
```

```python
classes = {
    0:"Warning sign",
    1:"Priority sign",
    2:"Forbidding sign",
    3:"Prescriptive sign",
    4:"special regulation sign",
    5:"Information sign",
    6:"Service Marks sign",
    7:"Additional Information signl"
    }
```

Preprocessing Image

```python
imgs = []
img= cv2.imread("/content/5_7001.png")
resized_img = cv2.resize(img, (100, 100))
imgs.append(resized_img)
array_img = np.array(imgs)
array_img.shape
```

```python
pred = model.predict(array_img)
```

```python
y_classes = [np.argmax(element) for element in pred]
```

```python
plt.imshow(resized_img)
print(classes[y_classes[0]])
```
